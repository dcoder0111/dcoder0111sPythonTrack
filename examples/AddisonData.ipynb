```python
number_types = []
for class_type in df['Class Type'].unique():
  number_types.append(df[df['Class Type']==class_type].count()[0])
print(number_types)
```

```python
data ={
        'types': df['Class Type'].unique(),
        'number_types': number_types
     }
  
new_df = pd.DataFrame(data)
```

```python
# To plot the waffle Chart
fig = plt.figure(
    title={
        'label': 'label',
        'loc': 'center',
        'fontdict': {
            'fontsize': 40
        }
    },
    figsize=(20,33),
    FigureClass = Waffle,
    rows = 5,
    colors=["#00AAAA", "#1155AA", "#EEEE01", "#99EEAA", "#FFA500", "#000000", "#EE333D", "#4444FF", "#CCCCCC"],
    values = new_df.number_types,
    labels = list(new_df.types),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1), 'fontsize': 30},
    interval_ratio_x = 0.5,
    interval_ratio_y = 0.5,
    icons='futbol',
    icon_size = 10,
    vertical=True,
    block_aspect_ratio=1,
    block_arranging_style='snake'
)
```
