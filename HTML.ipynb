```python
import pandas as pd
cat = 'basketball'
gender = 'woman'
url = f'https://www.nike.com.hk/{gender}/{cat}/shoe/list.htm?intpromo=PETP'
page = requests.get(url).content
soup_page = soup(page,'html')
shoes = soup_page.findAll('dl', {'class': 'product_list_content'})
shoe_names = []
shoe_prices = []
for shoe in shoes:
  shoe_names.append(shoe.findAll('span', {'class': 'up'})[0].text)
  shoe_prices.append(shoe.findAll('dd', {'class': 'color666'})[0].text)

dictionary = {
    'shoe_names' : shoe_names,
    'shoe_prices' : shoe_prices,
}
nike = pd.DataFrame(dictionary)
nike['current_prices'] = nike['shoe_prices'].apply(lambda x: x.split('HK$')[-1])
def removeCommas(x):
  if len(x) > 4:
    x = x[0] + x[2:]
  return x
nike['current_prices'] = nike['current_prices'].apply(removeCommas)
nike['current_prices'] = nike['current_prices'].apply(int)
```

```python
nike.to_csv('shoes.csv')
```
